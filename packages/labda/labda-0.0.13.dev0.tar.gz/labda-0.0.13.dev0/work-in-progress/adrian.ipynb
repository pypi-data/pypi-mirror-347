```python
import pandas as pd

df = pd.read_csv("../temp/days.csv")
activities = ["non_wear", "lie", "sit", "walk", "row", "mvpa", "vpa"]
df[activities] = df[activities].apply(lambda x: pd.to_timedelta(x, unit="min"))
df.set_index("datetime", inplace=True)
df[["non_wear", "lie", "sit", "walk", "row", "mvpa", "vpa"]] = df[
    ["non_wear", "lie", "sit", "walk", "row", "mvpa", "vpa"]
].apply(lambda x: x.dt.round("s"))
df.loc[df["id"] == 55000858, ["non_wear", "lie", "sit", "walk", "row", "mvpa", "vpa"]]
```
