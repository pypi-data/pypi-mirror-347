```python

# define and solve the test_DEAt model
def DEAweak_CRS():

    model = DEAweak.DEAweak2(data,sent = "K L E=Y:CO2", gy=[0],gx=[0,0,0],gb=[1], rts=RTS_CRS, baseindex=None,refindex=None)
    res = model.optimize(solver="mosek")
    # print(model.info(1))
    return res
DEAweak_CRS = DEAweak_CRS()
DEAweak_CRS
```

```python
def DEAweak_VRS1():


    model = DEAweak.DEAweak2(data,sent = "K L E=Y:CO2", gy=[0],gx=[0,0,0],gb=[1], rts=RTS_VRS1, baseindex=None,refindex=None)
    res = model.optimize(solver="mosek")
    # print(model.info(1))
    return res

DEAweak_VRS1 = DEAweak_VRS1()
DEAweak_VRS1
```
